```python
import pandas as pd
```

```python
df_population_raw = pd.read_csv("population_total.csv")
```

```python
df_population_raw.head(6)
```

```python
# droping nulls
df_population_raw.dropna(inplace=True)
```

```python
# MAKING A PIVOT TABLE
df_pivot = df_population_raw.pivot(index="year", columns="country",
                        values="population") 
```

```python
# notice we can adjust our new pivot so we can work (not visualize, but pivot)
df_pivot = df_pivot[["United States", "Brazil", "Chile", "France", "China", "South Africa"]]
```

```python
# showing pivot table
df_pivot
```

```python
## LINEPLOT

df_pivot.plot(kind = "line", xlabel="Year", ylabel="Population", title="Population Growth (1955-2020)",
              figsize=(8,3)) # kind = type of line, segmented, etc
```

```python
# BARPLOT

df_pivot.index # gives us data's index used for plotting
df_p_2010 = df_pivot[df_pivot.index.isin([2010])] # gives us data for an especific gap year
```

```python
df_p_2010 # showing new made-up dataframe
```

```python
# transpose dataframe from rows to columnns
df_p_2010_t = df_p_2010.T
```

```python
# making a barplot
df_p_2010_t.plot(kind='bar', color = 'c', xlabel="Year", ylabel="Population", title="Population Growth (1955-2020)")
```

```python
# barplot grouped by 'n' variables
df_pivot
df_p_sample = df_pivot[df_pivot.index.isin([1960, 1970, 1980, 1990, 2000])]
df_p_sample
```

```python
# grouped barplot
df_p_sample.plot(kind='bar')
```

```python
# PIECHART

df_p_2010_t.rename(columns={2010:'2010'}, inplace=True)
```

```python
df_p_2010_t.plot(kind='pie', y='2010', title="Population in 2010 (%)")
```

```python
# SAVE PLOT AND EXPORT PIVOT TABLE

import matplotlib.pyplot as plt

df_pivot.plot(kind = "line", xlabel="Year", ylabel="Population", title="Population Growth (1955-2020)",
              figsize=(8,3)) # kind = type of line, segmented, etc

# SAVE PLOT
plt.savefig('my_file.png') # Saved in the same folder as the code is

# SHOW PLOT
plt.show()
```
