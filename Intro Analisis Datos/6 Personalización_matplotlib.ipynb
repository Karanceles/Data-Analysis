```python
import matplotlib.pyplot as plt
```

```python
# Gráfico de barras
categorias = ["A", "B", "C", "D"]
valores = [10, 20, 25, 30]
plt.bar(categorias,valores) # bar graph
plt.title("Bar graphs")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
