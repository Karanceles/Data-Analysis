```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="darkgrid")

df = sns.load_dataset("tips")
df.head() # we only show the firsts rows of data
```

```python
df.info() # gaves us info, variables, n° rows, void or non-null data
```

```python
df.describe() # describes us data numerically, n° data, mean, deviations, prom, min max, etc
```

```python
# so we can visualize this variables using univariated analysis
sns.histplot(data=df, x="total_bill") # barlot
plt.plot()
```

```python
sns.boxplot(data=df, x="total_bill")
plt.plot()
```

```python
sns.kdeplot(data=df, x="total_bill")
plt.plot()
```

```python
# ANALISIS BIVARIADO (2 variado)
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

```python
sns.set_style("whitegrid")
palette = sns.color_palette("dark", 2)
sns.barplot(data=df, x="sex", y="tip", hue="sex", palette=palette)
plt.show()
```

```python
sns.boxplot(data=df, x="sex", y="total_bill", hue="sex", palette=palette)
plt.show()
```

```python
# ANALISIS MULTIVARIADO
sns.pairplot(df, hue="sex")
plt.show()
```
