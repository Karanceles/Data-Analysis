```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style="darkgrid")
palette = sns.color_palette("pastel", 4)
pal2 = sns.color_palette("dark", 2)
df = sns.load_dataset("tips")
df
```

```python
# bargraph of average tips
sns.barplot(data=df, x="sex", y="tip", hue="sex", errorbar=None, palette= pal2)
plt.title("average tips per sex")
plt.show()
```

```python
sns.barplot(data=df, x="day", y="tip", hue="day", errorbar=None, palette=palette)
# if I sleect hue as sex, for instance, the graph will be divided by tips and also sex
plt.title("average tips per day")
plt.show()
```

```python
sns.barplot(data=df, x="time", y="tip", hue="time", errorbar=None, palette=pal2)
plt.title("average tips per daytime")
plt.show()
```

```python
# smoker or not ?? afect tips ??
sns.barplot(data=df, x="smoker", y="tip", hue="day", errorbar=None, palette=palette)
plt.title("average tips by smokers")
plt.show()
```

```python
sns.barplot(data=df, x="day", y="tip", hue="sex", errorbar=None, palette=pal2)
plt.title("average tips by sex per day")
plt.show()
```

```python
sns.barplot(data=df, x="time", y="tip", hue="sex", errorbar=None, palette=pal2)
plt.title("average tips by time per sex")
plt.show()
```

```python
sns.barplot(data=df, x="smoker", y="tip", hue="sex", errorbar=None, palette=pal2)
plt.title("average tips by smokers per sex")
plt.show()
```

```python
# Dispersion graph in order to explore party attendants
sns.scatterplot(data=df, x="size", y="tip")
plt.title("Relations between attendant size and tips")
plt.show()
```

```python
# Bargraph calculating average tip per party size
palpal = sns.color_palette("pastel", 6)
sns.barplot(data=df, x="size", y="tip", hue="size" , errorbar=None, palette=palpal)
plt.title("average tips per party size")
plt.show()
```

```python
# calculate tips percentage, add to dataframe
df["tip_percentage"] = (df["tip"] / df["total_bill"] *100)
df
```

```python
sns.barplot(data=df, x="sex", y= "tip_percentage", hue="sex" , errorbar=None, palette=pal2)
plt.title("average tips percentage by sex")
plt.show()
```

```python
# bargraph by average tip percentage
sns.barplot(data=df, x="size", y= "tip_percentage", hue="sex" , errorbar=None, palette=pal2)
plt.title("average tips percentage by sex per size")
plt.show()
```

```python
sns.barplot(data=df, x="day", y= "tip_percentage", hue="day" , errorbar=None, palette=palette)
plt.title("average tips percentage by day")
plt.show()
```

```python
sns.barplot(data=df, x="time", y= "tip_percentage", hue="time" , errorbar=None, palette=pal2)
plt.title("average tips percentage by time")
plt.show()
```

```python
sns.barplot(data=df, x="smoker", y= "tip_percentage", hue="smoker" , errorbar=None, palette=pal2)
plt.title("average tips percentage by smoker")
plt.show()
```
