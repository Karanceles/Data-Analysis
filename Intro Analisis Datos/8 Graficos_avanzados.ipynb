```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# generate example data
np.random.seed(42) # setting a seed implies we can obtaing a certain randomized data
car_life = np.random.randint(0,20, size=200)
price_car = 30 - car_life + np.random.normal(-3, 3, size=200) # in case certain number is negative, change -3 per 0, so nothing is affected at least when a positive value is expected
data = pd.DataFrame({
    "Life" : car_life,
    "Price" : price_car
})
data
```

```python
# crafting a graph
fig, ax = plt.subplots(2, 2, figsize=(12,10))

# dispersioon graph // WHEN WORKING WITH AX OBJECTS, METHODS CHANGE
ax[0,0].scatter(data["Life"], data["Price"], alpha=0.5)
ax[0,0].set_xlabel("Car Lifetime (Years)")
ax[0,0].set_ylabel("Price (k USD)")
ax[0,0].set_title("Price-Life Distribution")

# Histograma de la edad en otro grafico (0,1)
sns.histplot(data["Life"], ax=ax[0,1],kde=True, color="c") # kde aproxima la distribucion
ax[0,1].set_xlabel("Car Lifetime (Years)")
ax[0,1].set_ylabel("Frequency")
ax[0,1].set_title("Histogram: Lifetime")

# Histograma del precio en otro grafico (1,0)
sns.histplot(data["Price"], ax=ax[1,0],kde=True, color="m") # kde aproxima la distribucion
ax[1,0].set_xlabel("Price (k USD)")
ax[1,0].set_ylabel("Frequency")
ax[1,0].set_title("Histogram: Price")

# Eliminate devoided subplots
ax[1,1].axis("off")

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```
