```python
# se busca cambiar la escala de los datos para visualizarlos de mejor manera

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
sns.displot(data=df, x = "sepal_length", hue="species", kde= True)
plt.show()
```

```python
# Log Transformation
df["sepal_length_log"] = np.log(df["sepal_length"])

sns.displot(data=df, x= "sepal_length_log", hue= "species", kde=True)
plt.show()
```
