```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
df = sns.load_dataset("iris")
df
```

```python
pca_model = PCA(n_components=2) # we reduced 4 components into 2 components
df_pca = pca_model.fit_transform(df.iloc[:,:-1]) # trains PCA model and returns the data transformed
df_pca

# .iloc[:,:-1] so we don't use species 'cause they don't have a numerical value
```

```python
# matplot to visualize the result
# this 'cause sns can't use an untransformed Data (This Numpy array has to be converted to a data frame)
plt.scatter(df_pca[:,0], df_pca[:,1], c=df["species"].astype("category").cat.codes)
# to colorize --> c=df["species"].astype("category").cat.codes
# c Array as colorlike
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()
 
```
