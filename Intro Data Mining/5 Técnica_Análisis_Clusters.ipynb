```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, random_state=42) # n_clusters or categories (versicolor, virginica, setosa)
# random state for an equal value to everyone
df = sns.load_dataset("iris")
df["cluster"] = model.fit_predict(df.iloc[:,:-1]) # new column
df
```

```python
# no longer hue = species but cluster
sns.scatterplot(data=df, x="sepal_length", y = "petal_length", hue = "cluster")
plt.show()
```
