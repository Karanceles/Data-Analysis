```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
# calcular el Z-Score
df = sns.load_dataset("iris")
df["z_score_length"] = np.abs((df["sepal_length"] - df["sepal_length"].mean()) / df["sepal_length"].std())
df
```

```python
# keep the outlayers
outlayers = df[df["z_score_length"] > 2]
outlayers # these are the ones that don't fit the model or distribution
```

```python
# inside our distribution
sns.scatterplot(data= df, x="sepal_length", y = "petal_length", hue = "species")

# outside our distribution // filtro con respecto a los datos
sns.scatterplot(data=outlayers,  x="sepal_length", y = "petal_length", color ="r")
```
